## Ejercicio 1. Sistema de empleados universitarios

Desarrolle un sistema orientado a objetos para representar personal universitario.

### Requisitos
- Cree una clase base llamada `Empleado` con:
  - `nombre`
  - `sueldo`
  - método `mostrar_datos()`
  - método `calcular_bono()`

- Cree dos clases hijas:
  - `Docente`
  - `Administrativo`

### Condiciones
- En `Docente`, agregue el atributo `horas_clase`.
- En `Administrativo`, agregue el atributo `departamento`.
- Sobrescriba el método `calcular_bono()` en cada clase hija.
- Use `super()` donde corresponda.
- Cree varios objetos y recórralos en una lista.

### Analice
- ¿Qué atributos y métodos deben estar en la clase padre?
- ¿Qué comportamiento cambia según el tipo de empleado?

## Ejercicio 2. Sistema de figuras con cálculo de área

Construya una jerarquía de clases para representar figuras geométricas.

### Requisitos
- Cree una clase base `Figura` con:
  - método `calcular_area()`
  - método `describir()`

- Cree las clases hijas:
  - `Rectangulo`
  - `Circulo`
  - `Triangulo`

### Condiciones
- Cada clase hija debe sobrescribir `calcular_area()`.
- Cada clase hija debe sobrescribir `describir()`.
- Recorra una lista de figuras y muestre:
  - el tipo de figura,
  - sus datos,
  - su área.

### Reto adicional
- Investigue cómo redondear resultados a 2 decimales.

## Ejercicio 3. Gestión de cuentas bancarias

Desarrolle un sistema de cuentas bancarias usando herencia.

### Requisitos
- Cree una clase base `CuentaBancaria` con:
  - `titular`
  - `saldo`
  - método `depositar(monto)`
  - método `retirar(monto)`
  - método `mostrar_info()`

- Cree dos clases hijas:
  - `CuentaAhorros`
  - `CuentaCorriente`

### Condiciones
- En `CuentaAhorros`, no permita retirar más dinero del saldo disponible.
- En `CuentaCorriente`, permita sobregiro hasta un valor límite.
- Sobrescriba el método `retirar()` en ambas clases hijas.
- Valide que no se ingresen montos negativos.
- Muestre mensajes claros al usuario.

### Analice
- ¿Por qué este problema requiere sobrescritura?
- ¿Qué reglas cambian de una cuenta a otra?

## Ejercicio 4. Inventario de dispositivos tecnológicos

  - `marca`

## Ejercicio 5. Plataforma académica con herencia multinivel

Modele una estructura académica con herencia multinivel.

### Requisitos
- Cree la clase `Persona`.
- Cree la clase `Estudiante` que herede de `Persona`.
- Cree la clase `EstudianteBecado` que herede de `Estudiante`.

### Condiciones
- `Persona` debe tener:
  - `nombre`
  - `cedula`

- `Estudiante` debe agregar:
  - `carrera`
  - `nivel`

- `EstudianteBecado` debe agregar:
  - `tipo_beca`
  - `porcentaje_descuento`

- Use `super()` correctamente en todos los niveles.
- Cree un método para mostrar toda la información.

### Analice
- ¿Qué atributos se heredan desde cada nivel?
- ¿Qué ventajas tiene esta jerarquía?
- ¿Qué desventajas tendría si crece demasiado?

## Ejercicio 6. Polimorfismo con reporte general

Desarrolle un sistema en el que distintos objetos respondan al mismo método.

### Requisitos
- Cree una clase base `UsuarioSistema`.
- Cree las clases hijas:
  - `Estudiante`
  - `Docente`
  - `Coordinador`

### Condiciones
- Todas las clases deben implementar el método `generar_reporte()`.
- Cada reporte debe mostrar información distinta según el rol.

### Reto adicional

## Ejercicio 7. Uso de `super()` en métodos no constructores

Construya un sistema donde `super()` se use también dentro de otros métodos.

### Requisitos
- Cree una clase base `Persona` con un método `presentarse()`.
- Cree una clase hija `Docente`.
- Cree una clase hija `Investigador`.

### Condiciones
- En las clases hijas, sobrescriba `presentarse()`.
- Use `super().presentarse()` para reutilizar parte del comportamiento.
- Luego agregue información específica en cada clase hija.

### Analice
- ¿Qué ventaja tiene extender un método en lugar de reemplazarlo completamente?

## Ejercicio 8. Reto integrador

Desarrolle un mini sistema orientado a objetos relacionado con un contexto real.

### Opciones
Puede elegir uno de estos contextos:
- sistema de transporte,
- sistema hospitalario,
