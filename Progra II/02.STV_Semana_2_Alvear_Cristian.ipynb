## ¿Qué es el encapsulamiento?

El **encapsulamiento** es un principio de la Programación Orientada a Objetos que permite:

- Proteger los datos internos de un objeto
- Controlar cómo se accede a esos datos
- Evitar modificaciones incorrectas desde el exterior
- Mejorar la seguridad y mantenibilidad del código

### Idea clave
Una clase no solo agrupa datos y comportamientos.  
También define **reglas** sobre cómo interactuar con sus atributos.

### Analogía
Piensa en un cajero automático:
- No accedes directamente al dinero interno
- Interactúas mediante botones, validaciones y reglas
- El sistema protege sus procesos internos

Con las clases ocurre algo similar.

````python
# Ejemplo simple sin encapsulamiento

class Persona:
    def __init__(self, nombre, edad):
        self.nombre = nombre
        self.edad = edad

p1 = Persona("Ana", 20) ## Aqui se a creado un objeto
print(p1.nombre, p1.edad)

````

````python
# Modificación directa
p1.edad = -5
print("Edad modificada directamente:", p1.edad)
````

````python
print(p1.edad)
````

## Problema del acceso directo

En el ejemplo anterior, el atributo `edad` puede modificarse libremente desde fuera de la clase.

Eso permite asignar valores inválidos, por ejemplo:

- edades negativas
- datos inconsistentes
- cambios sin validación

### Reflexión
¿Tiene sentido que cualquier parte del programa pueda alterar un dato importante sin control?

````python
# Clase con atributo privado

class Persona:
    def __init__(self, nombre, edad):
        self.nombre = nombre
        self.edad = edad

class PersonaProtegida:
    def __init__(self, nombre, edad):
        self.nombre = nombre
        self.__edad = edad

p2 = PersonaProtegida("Luis", 25)
print("Objeto creado correctamente")
````

````python
p2.__edad=18
````

````python
print(p2.__edad)
````

## Métodos de acceso

Cuando un atributo es privado, la forma correcta de interactuar con él es mediante métodos.

### Tipos comunes
- getter: permite consultar el valor
- setter: permite modificar el valor con validación

Esto ayuda a mantener coherencia en los datos.

````python
# Getter

class PersonaConGetter:
    def __init__(self, nombre, edad):
        self.nombre = nombre
        self.__edad = edad

    def obtener_edad(self):
        return self.__edad

p3 = PersonaConGetter("Carlos", 30)
print("Edad:", p3.obtener_edad())
````

## Setter con validación

Un setter permite cambiar el valor de un atributo privado, pero aplicando reglas.

### Ejemplo de validación
- la edad no debe ser negativa
- una nota debe estar entre 0 y 10
- un saldo no debería volverse inválido

````python
# Setter

class PersonaConSetter:
    def __init__(self, nombre, edad):
        self.nombre = nombre
        self.__edad = edad

    def obtener_edad(self):
        return self.__edad

    def cambiar_edad(self, nueva_edad):
        if nueva_edad >= 0:
            self.__edad = nueva_edad
            print("Edad actualizada correctamente")
        else:
            print("Error: la edad no puede ser negativa")

p4 = PersonaConSetter("María", 22)
print("Edad inicial:", p4.obtener_edad())

p4.cambiar_edad(23)
print("Edad actual:", p4.obtener_edad())

p4.cambiar_edad(-10)
print("Edad final:", p4.obtener_edad())
````

## Ejemplo aplicado: cuenta bancaria

El encapsulamiento es muy útil cuando trabajamos con datos sensibles.

### Caso
Una cuenta bancaria necesita:
- proteger el saldo
- permitir depósitos válidos
- evitar retiros inconsistentes
- mantener el estado correcto del objeto

````python
class CuentaBancaria():
    def __init__(self,titular,saldo):
        self.titular=titular
        self.__saldo=saldo
    @property
    def saldo(self):
        return self.__saldo
    @saldo.setter
    def saldo(self,nuevo_saldo):
        if nuevo_saldo>=0:
            self.__saldo=nuevo_saldo
````

### Instrucciones
Crear una clase llamada `Producto` con:

- atributo privado `__nombre`
- atributo privado `__precio`

### Métodos requeridos
- obtener nombre
- obtener precio
- actualizar precio
- mostrar información

### Regla
El precio solo debe actualizarse si es mayor que 0.

## **Código Cris Actividad Estudiante**

Definición de la Clase

````python
class Producto:
    def __init__(self, nombre, precio_inicial):
        # Atributos privados
        self.__nombre = nombre
        
        # Validación inicial del precio
        if precio_inicial > 0:
            self.__precio = precio_inicial
        else:
            self.__precio = 0.01 # Precio mínimo por defecto
            print("Aviso: El precio inicial debe ser mayor a 0.")

    # --- Métodos Getter (Obtener información) ---
    @property
    def nombre(self):
        return self.__nombre

    @property
    def precio(self):
        return self.__precio

    # --- Métodos Setter / Actualización ---
    @precio.setter
    def precio(self, nuevo_precio):
        # REGLA: El precio solo debe actualizarse si es mayor que 0
        if nuevo_precio > 0:
            self.__precio = nuevo_precio
            print(f"Precio de '{self.__nombre}' actualizado con éxito.")
        else:
            print("Error: El precio debe ser un valor positivo (mayor que 0).")

    def mostrar_informacion(self):
        print(f"\n[ DETALLES DEL PRODUCTO ]")
        print(f"Nombre: {self.__nombre}")
        print(f"Precio: ${self.__precio}")
        print("-" * 25)
````

Pruebas de Funcionamiento

````python
# 1. Crear el producto
articulo = Producto("Audífonos", 25.50)

# 2. Mostrar información inicial
articulo.mostrar_informacion()

# 3. Intentar actualizar a un precio inválido (Regla de negocio)
print("Intentando cambiar precio a -5.00...")
articulo.precio = -5.00 

# 4. Actualizar a un precio válido
print("\nIntentando cambiar precio a 29.99...")
articulo.precio = 29.99

# 5. Ver resultado final
articulo.mostrar_informacion()
````

## Ejercicio: Clase Vehiculo

Crea una clase llamada `Vehiculo`.

### Atributos
- `__marca`
- `__modelo`
- `__velocidad`

### Métodos
- obtener marca
- obtener modelo
- obtener velocidad
- acelerar
- frenar
- mostrar estado

### Reglas
- la velocidad no puede ser negativa
- no se puede frenar más allá de cero

### Prueba esperada
Simula varios cambios de velocidad.

## **CÓDIGO CRIS EJ VEHÍCULO**

Definición de la Clase

````python
class Vehiculo:
    def __init__(self, marca, modelo):
        # Atributos privados: no se pueden tocar desde afuera
        self.__marca = marca
        self.__modelo = modelo
        self.__velocidad = 0  # El vehículo siempre inicia detenido

    # --- Getters (Para consultar datos) ---
    @property
    def marca(self):
        return self.__marca

    @property
    def modelo(self):
        return self.__modelo

    @property
    def velocidad(self):
        return self.__velocidad

    # --- Métodos de Acción (Lógica de negocio) ---
    def acelerar(self, incremento):
        if incremento > 0:
            self.__velocidad += incremento
            print(f"Acelerando... Nueva velocidad: {self.__velocidad} km/h")
        else:
            print("Error: El incremento debe ser positivo.")

    def frenar(self, decremento):
        # Regla: No se puede frenar más allá de cero (no hay velocidad negativa)
        if decremento > 0:
            if self.__velocidad - decremento >= 0:
                self.__velocidad -= decremento
            else:
                self.__velocidad = 0 # Si frena de más, simplemente se detiene
            print(f"Frenando... Velocidad actual: {self.__velocidad} km/h")
        else:
            print("Error: La cantidad a frenar debe ser positiva.")

    def mostrar_estado(self):
        print(f"\n>>> ESTADO DEL VEHÍCULO <<<")
        print(f"Auto: {self.__marca} {self.__modelo}")
        print(f"Velocidad: {self.__velocidad} km/h")
        print("-" * 25)
````

Creación del Objeto

````python
# Creamos el objeto
mi_auto = Vehiculo("Toyota", "Corolla")

# Probamos los Getters
print(f"Vehículo listo: {mi_auto.marca} {mi_auto.modelo}")
mi_auto.mostrar_estado()
````

Pruebas de Velocidad

````python
print("--- Simulando Conducción ---")
mi_auto.acelerar(50)  # Sube a 50
mi_auto.acelerar(30)  # Sube a 80
mi_auto.frenar(40)    # Baja a 40

# Prueba de regla: Frenar más allá de cero
print("\n--- Intento de frenado excesivo ---")
mi_auto.frenar(100)   # Debería quedar en 0, no en -60
````

````python
print("--- Intento de modificación externa ---")

# Esto NO cambia el atributo real del objeto, crea uno nuevo "falso"
mi_auto.__velocidad = 500 

print(f"Valor 'hacker' intentado: {mi_auto.__velocidad}")
print("¿Qué dice el sensor real del auto?")
mi_auto.mostrar_estado() # Verás que sigue en su valor real protegido
````
