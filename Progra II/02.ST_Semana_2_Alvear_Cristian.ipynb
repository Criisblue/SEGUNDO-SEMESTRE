## Ejercicio 1: Billetera digital
Crea una clase llamada `BilleteraDigital`.

### Atributos privados
- `__titular`
- `__saldo`
- `__pin`

### Métodos
- consultar saldo
- recargar saldo
- pagar
- cambiar pin
- mostrar información básica

### Reglas
- no se puede recargar un valor menor o igual a cero
- no se puede pagar más del saldo disponible
- el PIN debe tener exactamente 4 dígitos
- el PIN no debe mostrarse directamente

## **CÓDIGO CRIS**

_Definición de la Clase

```python
class BilleteraDigital:
    def __init__(self, titular, saldo_inicial, pin):
        # El doble guion bajo (__) hace que los atributos sean privados
        self.__titular = titular
        self.__saldo = saldo_inicial
        self.__pin = str(pin) 

    def consultar_saldo(self):
        print(f"Saldo actual: ${self.__saldo}")

    def recargar_saldo(self, monto):
        # Validación: No recargar valores negativos o cero
        if monto > 0:
            self.__saldo += monto
            print(f"Recarga exitosa. Has sumado ${monto}")
        else:
            print("Error: El monto a recargar debe ser mayor a cero.")

    def pagar(self, monto):
        # Validación: No gastar más de lo que hay
        if monto <= self.__saldo:
            self.__saldo -= monto
            print(f"Pago de ${monto} realizado con éxito.")
        else:
            print(f"Error: Saldo insuficiente. Solo tienes ${self.__saldo}")

    def cambiar_pin(self, nuevo_pin):
        # Validación: Exactamente 4 dígitos
        pin_texto = str(nuevo_pin)
        if len(pin_texto) == 4 and pin_texto.isdigit():
            self.__pin = pin_texto
            print("PIN actualizado con éxito.")
        else:
            print("Error: El PIN debe tener exactamente 4 números.")

    def mostrar_informacion_basica(self):
        # Regla: Ocultar el PIN real
        print(f"--- Billetera de {self.__titular} ---")
        print(f"Saldo disponible: ${self.__saldo}")
        print(f"PIN: **** (Protegido)")
```

Creación del Objeto (Instancia)

```python
# Creamos la billetera de prueba
# Nombre: "Juan", Saldo inicial: 100, PIN: 1234
mi_billetera = BilleteraDigital("Juan Perez", 100, "1234")

# Probamos ver la info inicial
mi_billetera.mostrar_informacion_basica()
```

Operaciones de Dinero (Recargas y Pagos)

```python
print("--- Probando Recargas ---")
mi_billetera.recargar_saldo(50)   # Recarga correcta
mi_billetera.recargar_saldo(-10)  # Error: valor negativo

print("\n--- Probando Pagos ---")
mi_billetera.pagar(30)   # Pago correcto
mi_billetera.pagar(500)  # Error: saldo insuficiente

print("\n--- Saldo Final ---")
mi_billetera.consultar_saldo()
```

## Ejercicio 2: Cuenta de streaming
Crea una clase llamada `CuentaStreaming`.

### Atributos privados
- `__usuario`
- `__plan`
- `__meses_activos`
- `__estado_pago`

### Métodos
- obtener plan
- cambiar plan
- registrar pago
- suspender cuenta
- mostrar estado de la suscripción

### Reglas
- solo se permiten planes: Básico, Estándar y Premium
- no se puede cambiar a un plan inexistente
- no se debe exponer directamente el estado interno de pago

## **CÓDIGO CRIS EJ 2**

Definición de la Clase

```python
class CuentaStreaming:
    def __init__(self, usuario, plan_inicial):
        # Atributos privados (__). 
        # El estado de pago será un Booleano (True/False)
        self.__usuario = usuario
        self.__meses_activos = 0
        self.__estado_pago = False  # Inicia sin pagar
        
        # Validamos el plan al crear la cuenta
        planes_validos = ["Básico", "Estándar", "Premium"]
        if plan_inicial in planes_validos:
            self.__plan = plan_inicial
        else:
            print("Plan no válido. Se asignó 'Básico' por defecto.")
            self.__plan = "Básico"

    def obtener_plan(self):
        # Retorna el nombre del plan actual
        print(f"El plan actual de {self.__usuario} es: {self.__plan}")

    def cambiar_plan(self, nuevo_plan):
        # Regla: solo se permiten planes: Básico, Estándar y Premium
        planes_validos = ["Básico", "Estándar", "Premium"]
        if nuevo_plan in planes_validos:
            self.__plan = nuevo_plan
            print(f"Plan actualizado a: {self.__plan}")
        else:
            print(f"Error: El plan '{nuevo_plan}' no existe.")

    def registrar_pago(self):
        # Cambia el estado interno a pagado e incrementa los meses
        self.__estado_pago = True
        self.__meses_activos += 1
        print(f"¡Pago registrado! Gracias por renovar, {self.__usuario}.")

    def suspender_cuenta(self):
        # Pone el estado de pago en False
        self.__estado_pago = False
        print(f"La cuenta de {self.__usuario} ha sido suspendida.")

    def mostrar_estado_de_la_suscripcion(self):
        # Regla: no exponer directamente el estado interno de pago (__estado_pago)
        # En lugar de mostrar True/False, mostramos un mensaje amigable
        estado_texto = "Al día (Activo)" if self.__estado_pago else "Pendiente de pago (Inactivo)"
        
        print(f"--- Estado de Suscripción ---")
        print(f"Usuario: {self.__usuario}")
        print(f"Plan: {self.__plan}")
        print(f"Meses totales acumulados: {self.__meses_activos}")
        print(f"Estado: {estado_texto}")
        print("-----------------------------")
```

Creación y Validación Inicial

```python
# Crear una cuenta con un plan válido
cuenta_alex = CuentaStreaming("Alex123", "Premium")

# Intentar obtener el plan
cuenta_alex.obtener_plan()

# Ver estado inicial (debería salir como pendiente de pago)
cuenta_alex.mostrar_estado_de_la_suscripcion()
```

Pagos y Cambios de Plan

```python
print("--- Gestionando Pagos ---")
cuenta_alex.registrar_pago() # Mes 1
cuenta_alex.registrar_pago() # Mes 2

print("\n--- Intentando cambiar de Plan ---")
cuenta_alex.cambiar_plan("Estándar")  # Cambio válido
cuenta_alex.cambiar_plan("Ultra-HD") # Regla: plan inexistente (dará error)

cuenta_alex.mostrar_estado_de_la_suscripcion()
```

## Ejercicio 3: Sensor de temperatura inteligente
Crea una clase llamada `SensorTemperatura`.

### Atributos privados
- `__ubicacion`
- `__temperatura_actual`
- `__temperatura_minima`
- `__temperatura_maxima`

### Métodos
- registrar temperatura
- obtener temperatura actual
- obtener rango registrado
- mostrar estado del sensor

### Reglas
- cada nueva temperatura debe actualizar mínimos y máximos
- no se deben aceptar valores fuera de un rango razonable que tú definas
- la temperatura actual no debe modificarse directamente desde fuera de la clase

## **CÓDIGO CRIS EJ 3**

Definición de la Clase

```python
class SensorTemperatura:
    def __init__(self, ubicacion, temp_inicial):
        # Atributos privados
        self.__ubicacion = ubicacion
        self.__temperatura_actual = temp_inicial
        
        # Al iniciar, la actual es tanto la mínima como la máxima
        self.__temperatura_minima = temp_inicial
        self.__temperatura_maxima = temp_inicial

    def registrar_temperatura(self, nueva_temp):
        # Regla: No aceptar valores fuera de un rango razonable (-50 a 100)
        if -50 <= nueva_temp <= 100:
            self.__temperatura_actual = nueva_temp
            
            # Regla: Actualizar mínimos y máximos
            if nueva_temp < self.__temperatura_minima:
                self.__temperatura_minima = nueva_temp
            
            if nueva_temp > self.__temperatura_maxima:
                self.__temperatura_maxima = nueva_temp
            
            print(f"Temperatura registrada: {nueva_temp}°C")
        else:
            print(f"Error: {nueva_temp}°C está fuera del rango razonable.")

    def obtener_temperatura_actual(self):
        # Regla: La temp actual no debe modificarse desde fuera, solo leerse
        return self.__temperatura_actual

    def obtener_rango_registrado(self):
        # Devuelve los extremos históricos
        print(f"Rango histórico en {self.__ubicacion}:")
        print(f"Mínima: {self.__temperatura_minima}°C | Máxima: {self.__temperatura_maxima}°C")

    def mostrar_estado_del_sensor(self):
        print(f"--- Sensor: {self.__ubicacion} ---")
        print(f"Actual: {self.__temperatura_actual}°C")
        print(f"Estado: Funcionando correctamente")
```

Creación del Sensor y Lectura Inicial

```python
# Creamos un sensor en la "Cocina" iniciando a 20 grados
sensor_cocina = SensorTemperatura("Cocina", 20)

# Verificamos la lectura inicial
print(f"Lectura inicial: {sensor_cocina.obtener_temperatura_actual()}°C")
```

Registro de Temperaturas y Récords

```python
print("--- Registrando nuevas lecturas ---")
sensor_cocina.registrar_temperatura(25) # Nueva máxima
sensor_cocina.registrar_temperatura(18) # Nueva mínima
sensor_cocina.registrar_temperatura(22) # Temperatura intermedia

# Verificamos el rango histórico
sensor_cocina.obtener_rango_registrado()
```

Validación de Rangos Erróneos

```python
print("--- Probando límites de seguridad ---")
# Intentamos registrar un valor extremo (ej. 500 grados)
sensor_cocina.registrar_temperatura(500) 

# Intentamos registrar un valor extremadamente frío
sensor_cocina.registrar_temperatura(-100)

# Mostramos el estado final para confirmar que los valores locos se ignoraron
sensor_cocina.mostrar_estado_del_sensor()
```
