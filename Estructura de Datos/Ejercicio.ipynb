EJERCICIO DE ESTRUCTURA DE DATOS
LIMPIEZA DE CADENAS


Crea cadena, se calcula longitud

```python
raw_data="    1712345678, jUAN pablo Perez; Ingeniería de Software- Quito    "
print(f"longitud inicial de la cadena: {len(raw_data)}")
print(f"Cadena: {raw_data}")
  
```

*Eliminar espacios*

lstrip -  borra solo espacios de la izquierda

rstrip - borra solo espacios de la derecha


```python
paso1=raw_data.strip()
print(f"Nueva longitud {len(paso1)}")
print(f"nueva cadena {paso1}")
```

SACAR SOLO CÉDULA

SUBCADENA

```python
cedula=paso1[0:10]
print(f"Cédula: {cedula}")
```

Estandarizar los separadores

reemplazar los separadores

replace

```python
paso2=paso1.replace(";",",").replace("-",",")
print(f"Cadena nueva 2 {paso2}")
```

Estandarizar el NOMBRE

NORMALIZACIÓN DE TEXTOS


```python
paso3=paso2.title()
print(paso3)
```

Tratar a todos los datos por separado, romper la caderan y darle estructura

SPLIT

```python
cadena_final=paso3.split(",")
print(cadena_final)
```

AHORA TODOS SON TRATADOS POR SEPARADO

CÉDULA ES 0 

NOMBRES ES 1

ETC


```python
print(f"cédula {cadena_final[0]}")
print(f"Nombres {cadena_final[1]}")
print(f"Carrera {cadena_final[2]}")
print(f"Ciudad {cadena_final[3]}")
```
